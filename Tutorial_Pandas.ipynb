#  Introducción a **Pandas**

##  ¿Qué es Pandas?

**Pandas** es una **biblioteca de Python** diseñada para trabajar con **conjuntos de datos** (*datasets*).
Permite **analizar, limpiar, explorar y transformar datos** de forma sencilla, rápida y muy poderosa.

El nombre *Pandas* proviene de **“Panel Data”** y **“Python Data Analysis”**.
Fue creada por **Wes McKinney** en **2008** y se ha convertido en una herramienta esencial en ciencia de datos y aprendizaje automático.

---

## ¿Por qué usar Pandas?

En proyectos de **Inteligencia Artificial** y **Machine Learning**, los datos son la base de todo.
Sin datos limpios y organizados, ningún modelo funcionará correctamente.

Pandas nos ayuda a:

*  **Limpiar datos desordenados**: eliminar valores nulos, duplicados o inconsistentes.
*  **Explorar la información**: descubrir patrones, calcular estadísticas o relaciones entre variables.
*  **Preparar los datos para el aprendizaje automático**: normalizar, dividir o combinar conjuntos de datos.

 En resumen: **Pandas convierte datos en conocimiento**.

---

##  Conexión con la Ciencia de Datos

La **Ciencia de Datos** (*Data Science*) es una rama de la informática que estudia **cómo almacenar, usar y analizar datos para obtener información útil**.

Pandas es una de las herramientas fundamentales de esta disciplina, junto con **NumPy**, **Matplotlib** y **Scikit-learn**.

---

##  ¿Qué puede hacer Pandas?

Con Pandas puedes responder preguntas importantes sobre tus datos, por ejemplo:

* ¿Existe **correlación** entre dos o más columnas?
* ¿Cuál es el **valor medio, máximo o mínimo**?
* ¿Qué filas contienen datos incorrectos o vacíos?
* ¿Cómo puedo **filtrar, ordenar o combinar** distintas tablas?

Además, Pandas puede **eliminar o sustituir valores vacíos**, tarea conocida como **limpieza de datos** (*data cleaning*).

#  **Pandas Series**

##  ¿Qué es una *Series*?

Una **Series** de Pandas es como una **columna de una tabla** o de una hoja de cálculo.
Es una **estructura unidimensional** (una sola dimensión) que puede almacenar **datos de cualquier tipo**: números, cadenas, valores booleanos, etc.

Puedes imaginarla como una lista con nombre o una columna de Excel con índices.

---

##  Crear una *Series* desde una lista

📘 **Ejemplo:**

````python
import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
````

Cada valor tiene una **etiqueta o índice** (label).
Por defecto, Pandas asigna índices numéricos empezando desde 0.

---

##  **Índices o etiquetas (Labels)**

Si no especificas nada, los elementos se etiquetan automáticamente como:

```
0, 1, 2, 3, ...
```

Estos índices te permiten **acceder a un valor concreto**.

📘 **Ejemplo: acceder al primer valor**

````python
print(myvar[0])
````

---

##  **Crear tus propias etiquetas**

Puedes asignar tus **propias etiquetas** (índices personalizados) con el argumento `index`.

📘 **Ejemplo:**

````python
import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)
````


Y ahora puedes acceder por etiqueta en lugar de por posición:

````python
print(myvar["y"])
````

Otra forma muy útil de crear una Series es a partir de un **diccionario de Python**.
Las **claves** del diccionario se convierten automáticamente en los **índices**.

📘 **Ejemplo:**

````python
import pandas as pd

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
````

##  **Seleccionar elementos concretos**

Puedes elegir **solo algunas claves del diccionario** indicando los nombres que quieres conservar en el argumento `index`.

📘 **Ejemplo:**

````python
import pandas as pd

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day1", "day2"])

print(myvar)
````

 **De Series a DataFrame**

En Pandas, los conjuntos de datos suelen tener **varias columnas** (no solo una).
Cuando combinamos varias *Series* con distintos nombres de columna, obtenemos un **DataFrame**, que es una **tabla completa de datos**.

📘 **Ejemplo:**

````python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

myvar = pd.DataFrame(data)

print(myvar)
````

En este ejemplo, **cada columna es una Series**, y el **DataFrame** las une en una sola estructura.

---

#  Introducción a los DataFrames en Pandas

## ¿Qué es un DataFrame?

Un **DataFrame** de **Pandas** es una estructura de datos **bidimensional** (2D), similar a una **tabla** con filas y columnas.
Piensa en él como una hoja de cálculo de Excel, pero **dentro de Python**, con superpoderes para análisis de datos.

🔹 Cada columna puede tener un tipo distinto (números, texto, booleanos…).

🔹 Cada fila tiene un **índice** (por defecto empieza en 0).

🔹 Es la base del **análisis de datos** y de la **preparación de datasets** en IA.

---

##  Crear un DataFrame


````python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

# Cargar los datos en un objeto DataFrame
df = pd.DataFrame(data)

print(df)
````

Observa:

* Cada clave del diccionario (`"calories"`, `"duration"`) se convierte en una **columna**.
* Cada elemento de la lista es un **valor** de fila.

---

## Localizar filas con `loc`

Puedes acceder a filas específicas usando el **índice** o el **nombre de la fila** (si lo has definido).

### Ejemplo 1: Obtener la primera fila

````python
print(df.loc[0])
````

**Devuelve una Series**, no un DataFrame, porque solo hay una fila.

---

📘 **Ejemplo 2: devolver varias filas**

````python
print(df.loc[[0, 1]])
````

En este caso, al usar una **lista de índices**, Pandas devuelve otro **DataFrame**.

---

## 🏷️ **Índices con nombre (Named Indexes)**

Puedes asignar **nombres personalizados** a las filas mediante el argumento `index`.

📘 **Ejemplo:**

````python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index=["day1", "day2", "day3"])

print(df)
````

Ahora las filas tienen **etiquetas significativas** en lugar de números.

---

## 📍 **Acceder a una fila por nombre**

Una vez asignados nombres, puedes usar `.loc[]` con esos nombres.

📘 **Ejemplo:**

````python
print(df.loc["day2"])
````

#  Lección: Leer archivos CSV con Pandas

##  ¿Qué es un archivo CSV?

Los archivos **CSV** (*Comma Separated Values*) son una forma **sencilla y universal** de guardar grandes conjuntos de datos.
Cada línea representa una **fila**, y las columnas están separadas por **comas (,)**.

🔹 Son archivos de texto plano.
🔹 Se pueden abrir con Excel, Google Sheets o cualquier editor de texto.
🔹 Se usan constantemente en proyectos de **Big Data** y **Machine Learning**.

---

##  Cargar un archivo CSV con Pandas

Pandas facilita enormemente leer un archivo CSV.
Vamos a cargar un dataset llamado `data.csv`.

````python
import pandas as pd

url = "https://raw.githubusercontent.com/MainakRepositor/Datasets/refs/heads/master/Breast%20Cancer%20Dataset.csv"

df = pd.read_csv(url)

print(df.to_string())
````


 `to_string()` muestra **todo el DataFrame completo** (aunque sea grande).

---

###  Sin `to_string()`

Si no usamos `to_string()`, Pandas mostrará **solo las primeras y últimas 5 filas**, para no saturar la pantalla:

````python
import pandas as pd

url = "https://raw.githubusercontent.com/MainakRepositor/Datasets/refs/heads/master/Breast%20Cancer%20Dataset.csv"

df = pd.read_csv(url)

print(df)
````


---

## ⚙️ Controlar cuántas filas se muestran

Pandas tiene una configuración interna llamada `pd.options.display.max_rows`, que define el número máximo de filas que se muestran por defecto.

---

### 🔍 Ver cuántas filas máximas se muestran

````python
import pandas as pd

print(pd.options.display.max_rows)
````


➡️ Esto significa que, si tu DataFrame tiene más de 60 filas, solo se mostrarán **las primeras y las últimas 5**.

---

### 🔧 Cambiar el límite de filas visibles

Si quieres mostrar **todas** las filas de un DataFrame, puedes aumentar este límite:

````python
import pandas as pd

pd.options.display.max_rows = 9999   # Aumentamos el máximo

url = "https://raw.githubusercontent.com/MainakRepositor/Datasets/refs/heads/master/Breast%20Cancer%20Dataset.csv"

df = pd.read_csv(url)

print(df)
````


⚠️ **Cuidado:** Si tu archivo tiene miles de filas, puede tardar en mostrarse todo.

Supón que tienes una API que devuelve datos meteorológicos en formato JSON.
Con Pandas puedes cargar esos datos directamente y analizarlos.

````python
pip install requests
````

````python
import requests
import pandas as pd

# 1. Definir los parámetros para la API
# Coordenadas de León, España
latitud = 42.60
longitud = -5.57

# Rango de fechas para los datos históricos (ej. todo el año 2023)
fecha_inicio = "2023-01-01"
fecha_fin = "2023-12-31"

# Variables diarias que queremos obtener:
# - temperature_2m_max: Temperatura máxima diaria a 2 metros del suelo.
# - temperature_2m_min: Temperatura mínima diaria.
# - precipitation_sum: Suma total de precipitación (lluvia, nieve...) en mm.
# - wind_speed_10m_max: Velocidad máxima del viento a 10 metros en km/h.
parametros_diarios = "temperature_2m_max,temperature_2m_min,precipitation_sum,wind_speed_10m_max"

# 2. Construir la URL para la API de datos históricos
# ¡OJO! La URL es diferente a la de pronóstico: usa "archive-api.open-meteo.com"
url = (
    f"https://archive-api.open-meteo.com/v1/archive?"
    f"latitude={latitud}&longitude={longitud}"
    f"&start_date={fecha_inicio}&end_date={fecha_fin}"
    f"&daily={parametros_diarios}"
    f"&timezone=Europe/Madrid"  # Especificar la zona horaria es una buena práctica
)

try:
    # 3. Realizar la petición GET a la API
    respuesta = requests.get(url)
    respuesta.raise_for_status()  # Lanza un error si la petición no fue exitosa (ej. error 4xx o 5xx)

    # 4. Convertir la respuesta a formato JSON
    datos_json = respuesta.json()

    # 5. Extraer los datos diarios y guardarlos en un DataFrame de pandas
    datos_diarios = datos_json.get('daily')

    if datos_diarios:
        df_historico = pd.DataFrame(datos_diarios)

        # Convertir la columna 'time' a un formato de fecha más manejable
        df_historico['time'] = pd.to_datetime(df_historico['time'])

        # Renombrar columnas para mayor claridad
        df_historico.rename(columns={
            'time': 'Fecha',
            'temperature_2m_max': 'Temp Max (°C)',
            'temperature_2m_min': 'Temp Min (°C)',
            'precipitation_sum': 'Precipitación (mm)',
            'wind_speed_10m_max': 'Viento Max (km/h)'
        }, inplace=True)

        print(f"--- Datos meteorológicos históricos de León para 2023 ---")
        print(f"Se han obtenido {len(df_historico)} registros.")

        print("\nPrimeros 5 días del año:")
        print(df_historico.head(50))

        print("\nÚltimos 5 días del año:")
        print(df_historico.tail())

    else:
        print("No se encontraron datos diarios en la respuesta de la API.")

except requests.exceptions.RequestException as e:
    print(f"Error al conectar con la API: {e}")

````

## 🧩 **3. Sustituir un valor erróneo (método directo)**

Podemos **corregir manualmente** un valor aislado con `.loc[]`:

````python
df.loc[7, "Duration"] = 45

print(df.head(10))
````


✅ Esto reemplaza el valor erróneo por **45** en la fila 7.

📌 Útil cuando el dataset es pequeño o los errores son muy concretos.

---

##  **4. Crear reglas automáticas para detectar y corregir**

En datasets grandes no podemos corregir valor a valor.
Podemos definir **límites válidos (umbrales)** y aplicar reglas automáticas.

Por ejemplo, si todas las duraciones deben estar entre 30 y 120 minutos:

````python
url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/data.csv'
df = pd.read_csv(url)

for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.loc[x, "Duration"] = 120

print(df.head(10))
````


💡 `.loc[]` actúa como una lupa sobre el DataFrame y permite modificar datos directamente.

---

## 🔁 **5. Alternativa más “pandasiana” (sin bucles)**

Pandas permite hacerlo sin necesidad de recorrer fila a fila:

````python
url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/data.csv'
df = pd.read_csv(url)

df.loc[df["Duration"] > 120, "Duration"] = 120


print(df.head(10))
````


✅ Mismo resultado, pero más rápido y legible.

---

##  **6. Eliminar filas con valores erróneos**

En algunos casos no conviene reemplazar, sino **eliminar completamente** los registros que rompen la coherencia.

````python
url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/data.csv'
df = pd.read_csv(url)

df = df.drop(df[df["Duration"] > 120].index)

print(df.head(10))

````


💡 Esta opción se usa cuando:

* No sabemos cuál sería el valor correcto.
* Los datos erróneos son pocos y no afectan el análisis.

---

##  **7. Comprobar la corrección**



````python
print(df["Duration"].describe())
````


 Ya no hay valores anómalos por encima de 120.

    - Corregir es mejor cuando no quieres perder la información útil que sí tiene esa columna. Eliminar cuando el dato está completamente corrupto
2. ¿Qué consecuencias tendría mantener datos erróneos en un modelo de predicción?

## ✅ **Resumen**

#  **Eliminación de duplicados en Pandas**

---

##  **1. Qué son los duplicados**

En los datasets reales, es habitual encontrar **filas duplicadas**, es decir, **registros repetidos exactamente iguales**.

Esto puede suceder cuando:

* Se combinan fuentes de datos distintas.
* Se guardan varias veces las mismas mediciones.
* Hay errores de registro o importación.

🔹 **Ejemplo del dataset de entrenamientos**
Observa las filas 11 y 12: son idénticas.

````python
url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/data.csv'
df = pd.read_csv(url)

print(df.to_string())
````

---

##  **2. Detectar duplicados con `duplicated()`**

Pandas ofrece el método `duplicated()` para detectar filas repetidas.

````python
print(df.duplicated())
````


💡 **Interpretación:**

* `True` → la fila es **un duplicado**.
* `False` → la fila es **única**.

---

##  **3. Ver solo las filas duplicadas**

Para ver las filas duplicadas directamente:

````python
print(df[df.duplicated()])
````


---

##  **4. Eliminar duplicados con `drop_duplicates()`**

Para **eliminar filas repetidas**, usamos `drop_duplicates()`.

````python
df.drop_duplicates(inplace=True)
````


📌 Con `inplace=True` modificamos el DataFrame original directamente.
Sin ese argumento, el método devuelve una **copia** limpia.

---

## ⚙️ **5. Comprobación final**


````python
print(df.duplicated().sum())
````


✅ Ya no hay filas duplicadas.

---

##  **6. Eliminar duplicados según columnas específicas**

A veces queremos considerar duplicado solo si coinciden ciertas columnas (por ejemplo, mismas fechas y pulsaciones):

````python
url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/data.csv'
df = pd.read_csv(url)

df.drop_duplicates(subset=["Date", "Pulse"], inplace=True)

print(df.duplicated().sum())
````


Así mantenemos la fila más reciente o completa, y eliminamos las redundantes.

---

## 🧠 **7. Evitar duplicados al combinar datos**

Cuando unimos datasets con `concat()` o `merge()`, es buena práctica revisar después si se han duplicado registros:

````python
df1 = pd.DataFrame({
    "Nombre": ["Ana", "Luis", "María"],
    "Edad": [23, 30, 25]
})

df2 = pd.DataFrame({
    "Nombre": ["Luis", "María", "Pedro"],
    "Edad": [30, 25, 28]
})

df = pd.concat([df1, df2])

print(df)

df.drop_duplicates(inplace=True)

print("\n Eliminamos duplicados\n")

print(df)
````


💬 **Consejo:** Incorpora siempre un paso de verificación de duplicados en tu flujo de limpieza (*pipeline*).

---

## ✅ **Resumen**

| Tarea               | Método                             | Descripción                    |
| ------------------- | ---------------------------------- | ------------------------------ |
| Detectar duplicados | `df.duplicated()`                  | Devuelve `True/False` por fila |
| Mostrar duplicados  | `df[df.duplicated()]`              | Filtra las filas repetidas     |
| Eliminar duplicados | `df.drop_duplicates(inplace=True)` | Borra las filas duplicadas     |
| Basado en columnas  | `subset=["col1", "col2"]`          | Compara solo esas columnas     |



#  **Lección: Correlaciones de Datos en Pandas**

---

##  **1. ¿Qué es una correlación?**

En análisis de datos e inteligencia artificial, una **correlación** mide **cómo se relacionan dos variables**:
si aumentan o disminuyen juntas, o si no guardan relación alguna.

Pandas nos permite calcular correlaciones de forma muy sencilla con el método `.corr()`.

---

##  **2. Dataset de ejemplo**

Usaremos el dataset de entrenamientos `training_data.csv` con más datos:


````python
url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/training_data.csv'
df = pd.read_csv(url)

print(df.head())
````


---

##  **3. Calcular la correlación**

Pandas tiene un método muy útil: **`df.corr()`**, que calcula automáticamente las correlaciones entre todas las columnas numéricas del DataFrame.

````python
print(df.corr(numeric_only=True))
````


---

##  **4. Interpretación de los resultados**

Cada número indica **la fuerza y dirección de la relación** entre dos variables.
El valor varía de **-1 a 1**:

| Valor           | Interpretación                | Significado                      |
| --------------- | ----------------------------- | -------------------------------- |
| **1.0**         | Correlación perfecta positiva | Ambas variables aumentan juntas  |
| **0.8 a 0.6**   | Correlación fuerte            | Si una sube, la otra suele subir |
| **0.4 a 0.2**   | Correlación débil             | Relación poco fiable             |
| **0.0**         | Sin correlación               | No hay relación lineal           |
| **-0.6 a -1.0** | Correlación negativa          | Cuando una sube, la otra baja    |

---

##  **5. Ejemplo visual: buena vs mala correlación**

| Ejemplo                           | Descripción                                                  | Interpretación        |
| --------------------------------- | ------------------------------------------------------------ | --------------------- |
| 🔹 **Duration – Calories:** 0.92  | Cuanto más se entrena, más calorías se queman.               | Buena correlación ✅   |
| 🔹 **Duration – Maxpulse:** 0.009 | No hay relación directa entre la duración y el pulso máximo. | Mala correlación ❌    |
| 🔹 **Pulse – Maxpulse:** 0.78     | A mayor pulso medio, mayor pulso máximo.                     | Correlación fuerte 💪 |

---

##  **6. Visualización de correlaciones**

Podemos mostrar la matriz de correlaciones como un **mapa de calor (heatmap)** para interpretarla fácilmente.

````python
pip install Seaborn
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación de variables del entrenamiento")
plt.show()
````


 Esto genera un gráfico donde los colores indican la fuerza de la relación:

* 🔴 **Rojo fuerte → correlación positiva**
* 🔵 **Azul fuerte → correlación negativa**

---

## 💬 **7. Aplicación en Inteligencia Artificial**

En IA y *Machine Learning*, la correlación es fundamental porque:

* Permite **detectar variables redundantes o poco relevantes**.
* Ayuda a **reducir la dimensionalidad** del modelo (quedarse solo con las variables útiles).
* Sirve como paso previo a técnicas como **regresión lineal o PCA (análisis de componentes principales)**.

💡 Ejemplo:

> Si “Duration” y “Calories” tienen una correlación de 0.92, podrías usar una para predecir la otra con un modelo de regresión.
